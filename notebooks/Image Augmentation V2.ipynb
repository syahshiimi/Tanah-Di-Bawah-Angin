```python
print ('importing libraries')

import albumentations as A
import cv2
import os 
import glob
import numpy as np

from PIL import Image
from matplotlib import pyplot as plt

# if ready

print('libraries loaded...')
```

```python
# path configurations

images_path = '../content/shb/training_images/*.png'

export_path = '../content/shb/Albumentations/'

filename = 'shb_'
```

```python
# use glob to find image amount

for img in image_list = sorted(glob.glob(images_path))

print(img)

count = 1
```

```python
# use glob to create a sorted image list


for img in sorted(glob.glob(images_path)):
    
    # use compose feature to create an augmentation pipeline
               
    transform = A.Compose ([
        A.HorizontalFlip(p=0.5),
        A.VerticalFlip(p=0.2),
        A.RandomBrightness(p=1),
        A.GaussNoise(var_limit=(10.0, 50.0), p=0.5),
        A.MotionBlur(blur_limit=(3, 10), p=0.5),
        A.RandomContrast(limit=0.2, p=0.5),
        A.Resize(height=224, width=224, p=1)
    ])
    
    
    # use opencv to read images
    
    bgr_img = cv2.imread(img)
    rgb_image = cv2.cvtColor(bgr_img, cv2.COLOR_BGR2RGB)
    
    # run pipeline to the function A.Compose
    transformed = transform(image=rgb_image)
    # dictionary saved -> call to load image
    aug_img = transformed["image"] 
    # opencv reads in BGR, thus RGB from Albumentations must be converted
    aug_img = cv2.cvtColor(aug_img, cv2.COLOR_RGB2BGR) 
    
    # save images
    cv2.imwrite(export_path + filename + str(count) + '.jpg', aug_img) 
    count += 1
    cv2.waitKey(1)
    print(count)

```
